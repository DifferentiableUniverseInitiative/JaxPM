# Weak-lensing convergence: JaxPM Born vs GLASS

[![Open in Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/DifferentiableUniverseInitiative/JaxPM/blob/main/docs/notebooks/08-convergence-vs-glass.ipynb)

We run a JaxPM particle-mesh lightcone, build a full-sky HEALPix density shell at
each step, and compute the weak-lensing convergence $\kappa$ two ways:

1. **Density simulation** — LPT initial conditions evolved with a symplectic PM solver,
   painting one spherical shell per scale factor.
2. **Born convergence** — JaxPM's `convergence_Born` integral along the line of sight.
3. **GLASS convergence** — the same shells fed to `glass.MultiPlaneConvergence`.
4. **Map comparison** — $\kappa$ maps (`magma`) and the Born$-$GLASS residual (`RdBu_r`).
5. **Power spectra** — angular $C_\ell$ of the Born and GLASS maps compared.

Both pipelines use the *same* density shells and the *same* `jax_cosmo` distances, so the
Born/GLASS comparison isolates the lensing implementations themselves.

```python
import jax
import jax.numpy as jnp
import jax_cosmo as jc
import jax_cosmo.background as bg
import jax_cosmo.constants as const

import numpy as np
import healpy as hp
import matplotlib.pyplot as plt

from diffrax import ODETerm, SaveAt, diffeqsolve, SemiImplicitEuler, ConstantStepSize

from jaxpm.pm import linear_field, lpt
from jaxpm.ode import symplectic_ode
from jaxpm.lensing import spherical_density_fn, convergence_Born
from jaxpm.kernels import interpolate_power_spectrum

import glass

jax.config.update("jax_enable_x64", True)
```

## Configuration

```python
mesh_size = 400  # particle-mesh resolution (mesh_size^3)
nside = 256  # HEALPix resolution of the shells / maps
n_shells = 40  # number of spherical density shells
z_sources = [0.3, 0.5, 0.8]  # source redshifts for the convergence
seed = 42

# diffrax time stepping in scale factor
t0, t1, dt0 = 0.1, 1.0, 0.05

# Observer at the box centre -> the chi(z_max) sphere is inscribed in the box,
# giving a full-sky lightcone with no periodic replication.
observer_in_box = jnp.array([0.5, 0.5, 0.5])

mesh_shape = (mesh_size, mesh_size, mesh_size)
npix = hp.nside2npix(nside)
```

## Cosmology and geometry

```python
cosmo = jc.Planck18()

# The comoving distance to the furthest source sets the box: box = 2 * chi(z_max)
# so that chi(z_max) reaches exactly the box face (half the box) from the centre.
z_max = max(z_sources)
r_max = jc.background.radial_comoving_distance(cosmo, jc.utils.z2a(z_max)).squeeze()
box_size = tuple(float(2.0 * r_max) for _ in range(3))
observer_position = observer_in_box * jnp.array(box_size)

# Uniform shells in comoving distance.
r_edges = jnp.linspace(0.0, r_max, n_shells + 1)
r_center = 0.5 * (r_edges[1:] + r_edges[:-1])
a_center = jc.background.a_of_chi(cosmo, r_center)
d_R = float(r_max / n_shells)

print(f"box size       : {box_size[0]:.0f} Mpc/h")
print(f"chi(z={z_max}) : {float(r_max):.0f} Mpc/h")
print(f"shell width    : {d_R:.1f} Mpc/h over {n_shells} shells")
print(
    f"shell redshifts: z = {float(1 / a_center.max() - 1):.2f} .. {float(1 / a_center.min() - 1):.2f}"
)
```

## 1. Density simulation

LPT initial conditions are evolved with the symplectic PM solver. At each saved scale
factor we paint the particles that fall inside the corresponding shell onto a HEALPix map.

```python
# Linear power spectrum -> Gaussian initial conditions.
k = jnp.logspace(-3, 1, 256)
pk = jc.power.linear_matter_power(cosmo, k)
pk_fn = lambda x: interpolate_power_spectrum(x, k, pk)
ic = linear_field(mesh_shape, box_size, pk_fn, seed=jax.random.PRNGKey(seed))

# First-order LPT at the starting scale factor.
dx, p, _ = lpt(cosmo, ic, a=t0, order=1)

# Symplectic drift/kick operators (positions stored as displacements).
drift, kick = symplectic_ode(mesh_shape, cosmo, paint_absolute_pos=False)
terms = (ODETerm(kick), ODETerm(drift))

# Paint one HEALPix shell per saved scale factor (saved far -> near in a).
paint_shell = spherical_density_fn(
    mesh_shape, box_size, cosmo, nside, observer_position, d_R
)
saveat = SaveAt(ts=a_center[::-1], fn=lambda t, y, args: paint_shell(t, y[1], args))

sol = diffeqsolve(
    terms,
    SemiImplicitEuler(),
    t0=t0,
    t1=t1,
    dt0=dt0,
    y0=(p, dx),
    args=cosmo,
    saveat=saveat,
    stepsize_controller=ConstantStepSize(),
)

# Order the shells near -> far so they align with r_center / a_center.
density_planes = sol.ys[::-1]
print("density planes:", density_planes.shape)
```

## 2. Born convergence

`convergence_Born` integrates the lensing kernel over the shells. It converts each shell
to an overdensity internally, so it takes the raw painted density planes.

```python
z_arr = jnp.array(z_sources)
kappa_born = np.asarray(
    convergence_Born(cosmo, density_planes, r_center, a_center, z_arr, d_R)
)
print("Born kappa:", kappa_born.shape)
for z, kz in zip(z_sources, kappa_born):
    print(f"  z={z}: mean={kz.mean():+.2e}  rms={np.sqrt((kz**2).mean()):.4f}")
```

## 3. GLASS convergence

We feed the *same* shells to GLASS. To keep the geometry identical, GLASS reads distances
from a thin adapter around our `jax_cosmo` cosmology rather than a separate Boltzmann run.

```python
class JaxCosmoForGlass:
    """Expose a jax_cosmo cosmology through the GLASS Cosmology protocol.

    Distances are in Mpc/h; the h-factors cancel in GLASS's lensing ratios, so Born and
    GLASS see exactly the same geometry.
    """

    def __init__(self, cosmo):
        self._cosmo = cosmo

    @property
    def Omega_m0(self):
        return float(self._cosmo.Omega_m)

    @property
    def hubble_distance(self):
        return float(const.rh)

    def _chi(self, z):
        z = np.atleast_1d(np.asarray(z, dtype=np.float64))
        a = 1.0 / (1.0 + z)
        return np.asarray(
            jc.background.radial_comoving_distance(self._cosmo, jnp.array(a))
        )

    def transverse_comoving_distance(self, z, z2=None):
        chi = self._chi(z)
        if z2 is not None:
            chi = self._chi(z2) - chi
        return chi.item() if np.ndim(chi) == 0 else chi

    def H_over_H0(self, z):
        z = np.atleast_1d(np.asarray(z, dtype=np.float64))
        a = 1.0 / (1.0 + z)
        e = np.sqrt(np.asarray(bg.Esqr(self._cosmo, jnp.array(a))))
        return e.item() if np.ndim(e) == 0 else e


glass_cosmo = JaxCosmoForGlass(cosmo)
```

```python
# Overdensity per shell, in the same near -> far order as the Born planes.
deltas = [np.asarray(pl) / np.mean(np.asarray(pl)) - 1.0 for pl in density_planes]

# Top-hat windows from the shell edges in redshift.
z_edges = np.asarray(jc.utils.a2z(jc.background.a_of_chi(cosmo, r_edges)))
windows = glass.tophat_windows(z_edges)

# For each source redshift, accumulate every shell in front of it and place the
# source plane at the exact target redshift.
kappa_glass = []
for z_target in z_sources:
    mpc = glass.MultiPlaneConvergence(glass_cosmo)
    for delta_i, win in zip(deltas, windows):
        if win.zeff >= z_target:
            break
        mpc.add_window(delta_i, win)
    mpc.add_plane(np.zeros(npix), z_target, wlens=0.0)
    kappa_glass.append(np.copy(mpc.kappa))
kappa_glass = np.array(kappa_glass)
print("GLASS kappa:", kappa_glass.shape)
```

## 4. Map comparison

$\kappa$ maps in `magma`; the Born$-$GLASS residual in `RdBu_r` on a symmetric scale.
Born and GLASS share the same colour scale per redshift (clipped to the 1st/99th percentile
of the Born map) so the two methods can be compared directly by eye.

```python
nz = len(z_sources)

fig = plt.figure(figsize=(4.5 * nz, 6))
for i, z in enumerate(z_sources):
    # Shared, percentile-clipped scale so structure is visible and the two
    # methods are directly comparable (a few cluster pixels otherwise dominate).
    hp.mollview(
        jnp.log10(kappa_born[i] + 0.1),
        title=f"Born  z={z}",
        cmap="magma",
        sub=(2, nz, i + 1),
        cbar=True,
    )
    hp.mollview(
        jnp.log10(kappa_glass[i] + 0.1),
        title=f"GLASS  z={z}",
        cmap="magma",
        sub=(2, nz, i + 1 + nz),
        cbar=True,
    )
plt.show()

fig = plt.figure(figsize=(4.5 * nz, 3.2))
for i, z in enumerate(z_sources):
    diff = kappa_born[i] - kappa_glass[i]
    lim = 3.0 * diff.std()
    hp.mollview(
        diff,
        title=f"Born - GLASS  z={z}",
        cmap="RdBu_r",
        sub=(1, nz, i + 1),
        cbar=True,
        min=-lim,
        max=lim,
    )
plt.show()

for i, z in enumerate(z_sources):
    diff = kappa_born[i] - kappa_glass[i]
    rel = np.sqrt((diff**2).mean()) / np.sqrt((kappa_glass[i] ** 2).mean())
    corr = np.corrcoef(kappa_born[i], kappa_glass[i])[0, 1]
    print(f"z={z}: rel. RMS diff = {rel:.2%}, correlation = {corr:.4f}")
```

## 5. Power spectra

Angular power spectra of the two convergence maps, shown for $\ell \le 2\,N_\mathrm{side}$.
Born and GLASS use the same density shells and the same distances, so their spectra overlap
and their ratio (bottom panels) stays within ~1% across the whole band — the spectrum-level
confirmation that the two lensing implementations agree.

(No theory curve is shown: at $256^3$ the convergence power is shot-noise dominated relative
to a smooth prediction, so a theory comparison would not be informative at this resolution.)

```python
lmax = 2 * nside - 1
ell = np.arange(lmax + 1)
ellp = ell[2:]
coef = ellp * (ellp + 1) / (2.0 * np.pi)

fig, axes = plt.subplots(2, nz, figsize=(5 * nz, 8), sharex=True)
axes = np.atleast_2d(axes)
for i, z in enumerate(z_sources):
    cl_b = hp.anafast(kappa_born[i], lmax=lmax)[2:]
    cl_g = hp.anafast(kappa_glass[i], lmax=lmax)[2:]

    top = axes[0, i]
    top.loglog(ellp, coef * cl_b, label="Born (JaxPM)")
    top.loglog(ellp, coef * cl_g, ls="--", label="GLASS")
    top.set_title(f"z = {z}")
    top.grid(True, which="both", alpha=0.3)
    if i == 0:
        top.set_ylabel(r"$\ell(\ell+1)\,C_\ell^{\kappa\kappa}/2\pi$")
        top.legend()

    bot = axes[1, i]
    bot.semilogx(ellp, cl_b / cl_g, color="C0", label="Born / GLASS")
    bot.axhspan(0.98, 1.02, color="0.6", alpha=0.3, label=r"$\pm2\%$")
    bot.axhline(1.0, color="k", lw=0.8)
    bot.set_ylim(0.95, 1.05)
    bot.set_xlabel(r"$\ell$")
    bot.grid(True, which="both", alpha=0.3)
    if i == 0:
        bot.set_ylabel("Born / GLASS")
        bot.legend()

plt.tight_layout()
plt.show()
```
